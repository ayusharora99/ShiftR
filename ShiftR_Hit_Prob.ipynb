```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
from sklearn import preprocessing
```

```python
training_data = pd.read_csv("training_data.csv")
data_missing = training_data.isna()
data_num_missing = data_missing.sum()
data_num_missing/ len(training_data) * 100
```

```python
training_data = training_data.dropna(subset = ['exit_velocity','last_tracked_distance','hang_time','direction','play_result'])
```

```python
X = training_data[['exit_velocity', 'last_tracked_distance','hang_time','direction']]
```

```python
y = training_data['play_result']
```

```python
clf = LogisticRegression(random_state=42).fit(X, y)
```

```python
play_result_prob = clf.predict_proba(X)
```

```python
len(training_data)
```

```python
hit_probs = []
for probs in play_result_prob:
    hit_probs.append(probs[1])
```

```python
training_data['hit_prob'] = hit_probs
```
